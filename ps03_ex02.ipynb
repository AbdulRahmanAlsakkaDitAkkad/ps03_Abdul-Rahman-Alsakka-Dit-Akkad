```python
# Import necessary libraries for generating and displaying the word cloud
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
# Mount Google Drive to access files from Colab
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load the text file from Google Drive. Replace "path/to/your/text_file.txt" with your actual file path.
with open("/content/drive/My Drive/materials/JaneEyre.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
# Create a WordCloud object with specified parameters
wordcloud = WordCloud(
    width=800,             # Width of the word cloud image
    height=400,            # Height of the word cloud image
    background_color="white",  # Background color for the word cloud
    stopwords=set(STOPWORDS),  # Set of words to exclude from the word cloud
    min_font_size=10       # Minimum font size for the words in the word cloud
).generate(content)           # Generate the word cloud from the text
```

```python
# Display the generated word cloud using matplotlib
plt.figure(figsize=(8, 8), facecolor=None)  # Define figure size and background color
plt.imshow(wordcloud)       # Display the word cloud image
plt.axis("off")             # Hide axes
plt.tight_layout(pad=0)     # Remove padding around the word cloud
plt.show()
"""Observing and Analyzing the Word Cloud
Prominent Words: The word cloud will display words that appear frequently in the text as larger. This prominence is due to word frequency; common words are more visible.
Why: The importance of each word in the text reflects its frequency, helping identify key themes or ideas.
"""
```

```python
from PIL import Image
import numpy as np

# Load mask image
mask_image = np.array(Image.open("/content/drive/My Drive/materials/mask.png"))

# Create WordCloud with mask
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    stopwords=set(STOPWORDS),
    mask=mask_image,           # Use mask to shape the word cloud
    contour_width=1,
    contour_color='black'
).generate(content)

# Display word cloud with mask
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
